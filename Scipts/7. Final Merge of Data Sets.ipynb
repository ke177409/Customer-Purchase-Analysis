# 5. Checking and changing customers data types

```python
# assigning new data types by creating a dictionary

convert_dict = {'user_id': 'str',
                'Age': np.int32,
                'Number Dependents': np.int8
               }
```

```python
# overwriting old dtypes with new ones from dictionary

customers = customers.astype(convert_dict)
```

## Checking for mixed data types in customers df

```python
# First Name column contains mixed data types

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Changing First Name column data type to string

customers['First Name'] = customers['First Name'].astype('str')
```

```python
# verifying that change was made

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

# 9. Addressing outliers in prices column

```python
# Prices $14900 and 99999 are skewing results and are likely not correct
# Prices greater than $100 are labeled as NaN 

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# assigning new data types to ords_prods_merge by creating a dictionary

convert_dict2 = {'order_id': 'str',
                'user_id': 'str',
                'order_number': np.int32,
                'orders_day_of_week': np.int8,
                'order_hour_of_day': np.int8,
                'days_since_prior_order': np.float16,
                'product_id': 'str',
                'reordered': 'str',
                'aisle_id': 'str',
                'department_id': 'str',
                'prices': np.float32,
                'max_order': np.int32,
                'average_price': np.int32,
                'median_days_prior_order': np.float32
               }
```

# 11. Merging ords_prods_merge to customers

```python
# display headers to find key column

ords_prods_merge.head(1)
```

```python
# user_id is key column and is same data type as ords_prods_merge

customers.head(1)
```

```python
# inner join ords_prods_merge onto customers with an indicator flag
# merged data frame called full_merge

full_merge = ords_prods_merge.merge(customers, on = ['user_id'], indicator = True)
```

```python
# checking that merge was successful

full_merge.head()
```

```python
# verify that all rows were successfully merged

full_merge['_merge'].value_counts()
```

```python
#removing _merge flag column from final df

full_merge = ords_prods_merge.merge(customers, on = ['user_id'], indicator = False)
```

```python
# verify that values are accurate. AKA max price is $25

full_merge.describe()
```
